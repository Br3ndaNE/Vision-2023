```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import matplotlib.pylab as plt
import math
import numpy as np
```

```python
train, test = tfds.load(
    'stanford_dogs',
    split=['train', 'test'],
    batch_size=1,
    as_supervised=True,
)
```

```python
for img_batch, label_batch in train.take(5):  # Take one batch
    img = img_batch[0].numpy()  # Take the first image from the batch

# Visualize the image
print(label_batch)
plt.imshow(img)
plt.show()
```
